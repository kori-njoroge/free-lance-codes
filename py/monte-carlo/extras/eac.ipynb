```python
import numpy as np
import matplotlib.pyplot as plt

def apply_rules(grid):
    # create a copy of the grid
    new_grid = np.copy(grid)
    
    # apply the rules to each cell
    for i in range(1, grid.shape[0] - 1):
        for j in range(1, grid.shape[1] - 1):
            neighbors = [grid[i-1,j-1], grid[i-1,j], grid[i-1,j+1],
                         grid[i,j-1], grid[i,j], grid[i,j+1],
                         grid[i+1,j-1], grid[i+1,j], grid[i+1,j+1]]
            count_ones = neighbors.count(1)
            
            # apply the rules based on the counts
            if grid[i,j] == 0 and count_ones == 2:
                new_grid[i,j] = 1
            elif grid[i,j] == 1 and (count_ones < 2 or count_ones > 3):
                new_grid[i,j] = 0
    
    return new_grid

def classify_CA(grid):
    # get the initial density
    density = np.sum(grid) / grid.size
    
    # apply the rules for 50 iterations
    for i in range(50):
        grid = apply_rules(grid)
    
    # check the final state for each class
    if np.array_equal(grid, np.zeros(grid.shape)):
        return "Class 1"
    elif np.array_equal(grid, np.ones(grid.shape)):
        return "Class 2"
    elif density > 0.45 and density < 0.55:
        return "Class 3"
    else:
        return "Class 4"
        
def plot_CA(grid, title):
    plt.imshow(grid, cmap='binary')
    plt.title(title)
    plt.axis('off')
    plt.show()

# single cell seed
grid = np.zeros((100, 100))
grid[50, 50] = 1
for i in range(50):
    grid = apply_rules(grid)
plot_CA(grid, "Single Cell Seed")

# random seed
grid = np.random.randint(2, size=(100, 100))
for i in range(50):
    grid = apply_rules(grid)
classification = classify_CA(grid)
plot_CA(grid, f"Random Seed ({classification})")
```
