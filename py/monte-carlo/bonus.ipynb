```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cellular_automaton(rule_number):
    # Define the rule as a binary string of length 8
    rule_binary = format(rule_number, '08b')
    # Convert the binary string into a list of integers
    rule = [int(bit) for bit in rule_binary]
    # Define the initial state for the single cell seed
    N = 201
    single_cell_seed = np.zeros(N)
    single_cell_seed[N // 2] = 1

    # Define the size of the visualization
    t = 100
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
    for i in range(2):
        state = np.zeros((t+1, N))
        state[0, :] = single_cell_seed
        for n in range(t):
            for k in range(1, N-1):
                triplet = state[n, k-1:k+2].astype(int)
                state[n+1, k] = rule[int(''.join(map(str, triplet)), 2)]
        axs[i].imshow(state, cmap='Greys', interpolation='nearest')
        axs[i].set_title(f"Rule {rule_number}")
    plt.show()
```

```python
# texting the code
cellular_automaton(120)
```
